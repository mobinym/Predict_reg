````python
def clean_and_convert(df):
    """
    این متد را تکمیل کنید تا داده‌های ورودی را پاک‌سازی و تبدیل کند.

    وظایف مورد انتظار:
    1. تبدیل ستون 'Date' به فرمت datetime
       - ستون 'Date' را به فرمت `datetime` تبدیل کنید.
       - در صورت وجود خطا، مقدار نامعتبر را `NaT` قرار دهید.

    2. استخراج ساعت از ستون 'Time'
       - فرض کنید مقادیر این ستون به‌صورت 'hh.mm.ss' هستند.
       - مقدار ساعت را از این رشته استخراج کرده و به فرمت `float` تبدیل کنید.
       - نام ستون جدید را 'Hour' بگذارید.

    3. تبدیل مقادیر عددی در ستون‌های متنی
       - ابتدا، ستون‌هایی را که نوع داده آن‌ها `object` است شناسایی کنید.
       - بررسی کنید که آیا این ستون‌ها دارای مقدار عددی با کاما به‌جای نقطه هستند.
       - در صورت وجود، کاما را به نقطه تبدیل کنید و مقدار نهایی را به `float` تغییر دهید.

    در نهایت، دیتافریم اصلاح‌شده را بازگردانید.
    """
    
    # تبدیل ستون 'Date' به فرمت datetime
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    # استخراج ساعت از ستون 'Time' و ایجاد ستون 'Hour'
    df['Hour'] = df['Time'].apply(lambda x: float(x.split('.')[0]) + float(x.split('.')[1])/60 if isinstance(x, str) else None)

    # تبدیل مقادیر عددی در ستون‌های متنی
    for column in df.select_dtypes(include=['object']).columns:
        df[column] = df[column].str.replace(',', '.').astype(float, errors='ignore')

    # بررسی نوع داده‌ها پس از اعمال تغییرات
    print(df.dtypes)
    
    return df

# اعمال متد روی دیتاست
df_copy = clean_and_convert(df_copy)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import math

def plot_scatter_grid(df, target_variable, cols=3):
    """
    رسم نمودارهای پراکندگی برای هر ویژگی عددی در برابر متغیر هدف، در یک تصویر با چیدمان شبکه‌ای.

    پارامترها:
    - df: دیتافریم ورودی
    - target_variable: متغیر هدف برای محور Y
    - cols: تعداد ستون‌های گراف‌ها در تصویر (پیش‌فرض ۳)

    """
    if target_variable not in df.columns:
        raise ValueError(f"Target variable '{target_variable}' not found in the dataframe.")
    
    # استخراج ویژگی‌های عددی
    numerical_columns = df.select_dtypes(include='number').columns.tolist()
    
    # حذف متغیر هدف از لیست ویژگی‌ها
    if target_variable in numerical_columns:
        numerical_columns.remove(target_variable)
    
    # تنظیمات ساب‌پلات‌ها
    num_features = len(numerical_columns)
    rows = math.ceil(num_features / cols)  # محاسبه تعداد ردیف‌ها
    
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))  # تنظیم ابعاد نمودار
    axes = axes.flatten()  # یک بعدی کردن لیست محورها
    
    for i, column in enumerate(numerical_columns):
        sns.scatterplot(ax=axes[i], x=df[column], y=df[target_variable])
        axes[i].set_title(f'{column} vs {target_variable}')
        axes[i].set_xlabel(column)
        axes[i].set_ylabel(target_variable)
    
    # حذف نمودارهای اضافی در صورتی که تعداد ویژگی‌ها کمتر از تعداد ساب‌پلات‌ها باشد
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

# اعمال متد روی دیتاست
plot_scatter_grid(df_copy, 'CO(GT)')
````

````python
from scipy.stats.mstats import winsorize
import pandas as pd

def winsorize_and_handle_missing(df, cols, limits=[0.05, 0.05]):
    """
    این متد داده‌های گمشده را با مقادیر Winsorized جایگزین می‌کند.

    1. کپی‌برداری از دیتافریم:
       - یک کپی از دیتافریم ورودی ایجاد می‌شود تا تغییرات در دیتافریم اصلی ایجاد نشود.
    2. اعمال Winsorization:
       - برای هر ستون، از روش `winsorize` استفاده می‌شود تا داده‌ها را با مقادیر Winsorized در بازه‌های مشخص‌شده (با استفاده از آرگومان `limits`) جایگزین کنیم.
       - مقادیر کمتر از پایین‌ترین درصد (مثلاً 5%) و بالاتر از بالاترین درصد (مثلاً 95%) را به آن مقادیر تبدیل می‌کنیم.
    """
    # کپی‌برداری از دیتافریم
    df_copy = df.copy()

    # بررسی ستون‌های موجود در لیست cols
    for col in cols:
        if col in df_copy.columns:
            # اعمال Winsorization برای هر ستون
            df_copy[col] = winsorize(df_copy[col], limits=limits)
        else:
            print(f"Warning: Column '{col}' not found in the dataframe")

    return df_copy

# اعمال متد روی دیتاست
df_copy = winsorize_and_handle_missing(df_copy, cols=df_copy.select_dtypes(include='number').columns.tolist())
````

````python
# TODO : متدی بتویسید که همبستگی چند جمله ایی را بررسی و مدیریت کند (Multicolinearity)

import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

def remove_multicollinearity(df, threshold=10):
    """
    این تابع ویژگی‌هایی که همبستگی چندگانه دارند را بررسی کرده و ویژگی‌هایی با مقدار VIF بالا را حذف می‌کند.
    
    ورودی:
        df: دیتافریم شامل ویژگی‌های عددی
        threshold: مقدار آستانه VIF (پیش‌فرض 10)
    
    خروجی:
        دیتافریمی با ویژگی‌های بهینه‌شده و حذف ویژگی‌های دارای همبستگی بالا
    """
    df_copy = df.copy()
    
    # لیست ویژگی‌های عددی
    numeric_features = df_copy.select_dtypes(include=[np.number]).columns.tolist()
    
    while True:
        vif_data = pd.DataFrame()
        vif_data["Feature"] = numeric_features
        vif_data["VIF"] = [variance_inflation_factor(df_copy[numeric_features].values, i) for i in range(len(numeric_features))]
        
        # پیدا کردن ویژگی با بیشترین مقدار VIF
        max_vif = vif_data["VIF"].max()
        
        if max_vif < threshold:
            break  # اگر همه VIF ها کمتر از مقدار آستانه بودند، متوقف شود

        # حذف ویژگی با بالاترین VIF
        drop_feature = vif_data.loc[vif_data["VIF"].idxmax(), "Feature"]
        print(f"📌 حذف ویژگی: {drop_feature} (VIF={max_vif:.2f})")
        numeric_features.remove(drop_feature)
        df_copy.drop(columns=[drop_feature], inplace=True)

    return df_copy

# اعمال متد روی دیتافریم
df_optimized = remove_multicollinearity(df_copy)
````
